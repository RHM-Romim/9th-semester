```python
# CAESAR CIPHER IMPLEMENTATION
# A substitution cipher where each letter is shifted by a fixed 'key' value

def encrypt(text, key):
    result = ""
    
    for ch in text:  # Iterate over each character in the input text
        
        if ch.isupper():
            # For UPPERCASE letters:
            # 1. ord(ch) - ord('A')  → converts letter to 0-25 range  (e.g., 'A'=0, 'B'=1, 'Z'=25)
            # 2. + key               → shifts it forward by key positions (e.g., 'A'+3 = 3)
            # 3. % 26                → wraps around if it goes past 'Z'  (e.g., 25+3=28 → 28%26=2 → 'C')
            # 4. + ord('A')          → converts back to (ASCII 65) character
            result += chr((ord(ch) - ord('A') + key) % 26 + ord('A'))
        
        elif ch.islower():
            # Same logic as uppercase, but uses 'a' as the base (ASCII 97)
            result += chr((ord(ch) - ord('a') + key) % 26 + ord('a'))
        
        else:
            # Non-alphabetic characters (digits, spaces, symbols) are kept unchanged
            # e.g., '1', '2', '3', '4' stay as-is
            result += ch

    return result


def decrypt(text, key):
    # Decryption is just encryption with a NEGATIVE key
    # If encrypt shifts +3, decrypt shifts -3, reversing the operation
    return encrypt(text, -key)


# ─── DEMO ───────────────────────────────────────────────
plain_text = "My name is Romim"  # Original message
key = 3                    # Shift amount

print(f"Plaintext:           {plain_text}")

cipher_text = encrypt(plain_text, key)
print(f"Ciphertext:          {cipher_text}")   # Each letter shifted +3

decrypted_text = decrypt(cipher_text, key)
print(f"Decrypted Plaintext: {decrypted_text}")  # Should match original
```
