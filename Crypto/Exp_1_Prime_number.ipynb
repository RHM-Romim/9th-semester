```python
# ============================================================
# SIEVE OF ERATOSTHENES - Method 1 (Boolean Flag Array)
# Goal: Find all prime numbers from 2 to 100
# ============================================================

# STEP 1: Create a list of 101 elements (index 0 to 100),
# all initialized to 1 (meaning 'assume every number is prime')
arr = [1] * 101

# STEP 2: 0 and 1 are NOT prime numbers by definition, so mark them as 0
arr[0] = 0   # 0 is not prime
arr[1] = 0   # 1 is not prime

# STEP 3: Run the Sieve
# We only need to check up to sqrt(100) = 10, because:
# any composite number <= 100 must have a factor <= sqrt(100)
for i in range(2, int(100**0.5) + 1):  # i goes from 2 to 10

    if arr[i] == 1:   # if i is still marked as prime (not yet crossed out)

        # Mark all multiples of i as NOT prime (composite)
        # Start from i*i because smaller multiples (2i, 3i, ...) 
        # were already marked by previous primes
        # Step by i to hit every multiple of i
        for j in range(i * i, 101, i):
            arr[j] = 0   # cross out this multiple (it is composite, not prime)

# STEP 4: Print all numbers that are still marked as 1 (i.e., still prime)
print("Prime Numbers:", end=" ")
for i in range(2, 101):  # check every number from 2 to 100
    if arr[i] == 1:       # if the flag is still 1, it survived the sieve = prime
        print(i, end=" ")
```

```python
# ============================================================
# SIEVE OF ERATOSTHENES - Method 2 (Value-Based List)
# Same algorithm, but the list stores actual numbers instead of flags
# ============================================================

# STEP 1: Create a list [0, 1, 2, 3, ..., 100]
# Each index stores its own value, e.g., prime[5] = 5
prime = list(range(101))

# STEP 2: Mark 0 and 1 as non-prime by setting them to 0
prime[0] = 0   # 0 is not prime
prime[1] = 0   # 1 is not prime

# STEP 3: Run the Sieve
# Only iterate up to sqrt(100) = 10 (same reason as Method 1)
for i in range(2, int(100**0.5) + 1):  # i goes 2, 3, 4, ..., 10

    if prime[i] != 0:   # if this position hasn't been crossed out (still a prime candidate)

        # Eliminate all multiples of i by setting them to 0
        # Starting at i*i avoids redundant work (smaller multiples already handled)
        for j in range(i * i, 101, i):
            prime[j] = 0   # set composite number's slot to 0 (eliminated)

# STEP 4: Collect the surviving prime numbers into a clean list
only_primes = []   # this will hold just the prime numbers

for i in range(2, 101):
    if prime[i] != 0:          # if the value wasn't zeroed out, it's prime
        only_primes.append(prime[i])   # add it to the primes list

# STEP 5: Print the prime values
print("Value:", end=" ")
for x in only_primes:
    print(x, end=" ")   # print each prime number

print()   # newline after the values row

# STEP 6: Print the 1-based index (rank) of each prime
# e.g., 2 is the 1st prime, 3 is the 2nd, 5 is the 3rd, ...
print("Index:", end=" ")
for i in range(1, len(only_primes) + 1):
    print(i, end=" ")   # print the rank/position number
```
