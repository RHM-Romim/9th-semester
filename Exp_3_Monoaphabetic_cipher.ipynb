```python
# ============================================================
# MONOALPHABETIC CIPHER - Method 1 (Fixed Substitution Key)
# ------------------------------------------------------------
# A Monoalphabetic Cipher replaces each letter in the plaintext
# with a FIXED corresponding letter from a cipher alphabet.
# The mapping is always one-to-one and never changes (mono = one).
# Key = the cipher alphabet itself.
# ============================================================

import string

# The standard a-z alphabet — this is our plaintext reference
plain_alphabet = string.ascii_lowercase   # 'abcdefghijklmnopqrstuvwxyz'

# The cipher alphabet — each position maps to the same index in plain_alphabet
# e.g., plain 'a' (index 0) -> cipher 'q' (index 0)
#        plain 'b' (index 1) -> cipher 'w' (index 1), and so on
# This is the SECRET KEY shared between sender and receiver
cipher_alphabet = "qwertyuiopasdfghjklzxcvbnm"

# Substitution mapping (for reference):
# plain : a b c d e f g h i j k  l  m  n  o  p  q  r  s  t  u  v  w  x  y  z
# cipher: q w e r t y u i o p a  s  d  f  g  h  j  k  l  z  x  c  v  b  n  m


# ----------------------------------------------------------
# ENCRYPTION FUNCTION
# Replaces each plain letter with its cipher counterpart
# ----------------------------------------------------------
def encrypt(text):
    result = ""  # will hold the final encrypted string

    for char in text.lower():  # convert input to lowercase first

        if char in plain_alphabet:  # only substitute alphabetic characters

            # Find WHERE this character sits in the plain alphabet (its index)
            index = plain_alphabet.index(char)

            # Use the SAME index to pick the corresponding cipher letter
            result += cipher_alphabet[index]

        else:
            result += char  # non-alphabetic chars (spaces, digits) pass through unchanged

    return result


# ----------------------------------------------------------
# DECRYPTION FUNCTION
# Reverses the substitution: cipher letter -> plain letter
# ----------------------------------------------------------
def decrypt(text):
    result = ""  # will hold the recovered plaintext

    for char in text.lower():  # convert to lowercase

        if char in cipher_alphabet:  # only process cipher letters

            # Find WHERE this cipher character sits in the cipher alphabet
            index = cipher_alphabet.index(char)

            # Use the SAME index to get back the original plain letter
            result += plain_alphabet[index]

        else:
            result += char  # non-alphabetic chars pass through unchanged

    return result


# ----------------------------------------------------------
# MAIN: Take input, encrypt, then decrypt to verify
# ----------------------------------------------------------
text = input("Enter text: ")

encrypted = encrypt(text)    # encrypt the plain text
decrypted = decrypt(encrypted)  # decrypt the cipher text back to verify

print("Encrypted:", encrypted)
print("Decrypted:", decrypted)
```

```python
# ============================================================
# MONOALPHABETIC CIPHER - Method 2 (Random Shuffled Key)
# ------------------------------------------------------------
# Same concept as Method 1, but the cipher alphabet is
# RANDOMLY SHUFFLED each time the program runs.
# This makes it harder to guess the key (26! possible keys).
# Also handles UPPERCASE separately, preserving case.
# ============================================================

import random
import string

# ---- Setup the plain alphabets ----
plain_lower = string.ascii_lowercase   # 'abcdefghijklmnopqrstuvwxyz'
plain_upper = string.ascii_uppercase   # 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

# ---- Build the cipher key by randomly shuffling the alphabets ----
# Convert to list first so we can shuffle (strings are immutable in Python)
cipher_lower = list(plain_lower)   # ['a', 'b', 'c', ..., 'z']
cipher_upper = list(plain_upper)   # ['A', 'B', 'C', ..., 'Z']

# Randomly rearrange both lists in-place — this creates a random key
# Every run of the program generates a NEW random key
random.shuffle(cipher_lower)
random.shuffle(cipher_upper)

# Convert the shuffled lists back to strings for easy indexing
cipher_lower = "".join(cipher_lower)   # e.g., 'juoetscdhgxybpfzmlnvkraqwi'
cipher_upper = "".join(cipher_upper)   # e.g., 'ISBERTQNJVKGWXFMZUYDLPOAHC'

# Print the full substitution table so sender and receiver can see the key
print("Substitution Table:")
print("Plain :", plain_lower + plain_upper)    # original alphabet
print("Cipher:", cipher_lower + cipher_upper)  # scrambled (key) alphabet


# ----------------------------------------------------------
# ENCRYPTION FUNCTION (Case-Preserving)
# Lowercase letters -> substituted using cipher_lower
# Uppercase letters -> substituted using cipher_upper
# ----------------------------------------------------------
def encrypt(text):
    result = ""  # accumulate the encrypted output here

    for c in text:  # process each character one by one

        if c.islower():
            # Find the plain letter's position, use it to get cipher letter
            result += cipher_lower[plain_lower.index(c)]

        elif c.isupper():
            # Same logic but for uppercase letters
            result += cipher_upper[plain_upper.index(c)]

        else:
            result += c  # spaces, numbers, symbols are kept as-is

    return result


# ----------------------------------------------------------
# DECRYPTION FUNCTION (Case-Preserving)
# Reverses the substitution by looking up cipher letter in cipher alphabet
# and returning the corresponding plain letter
# ----------------------------------------------------------
def decrypt(text):
    result = ""  # accumulate the decrypted output here

    for c in text:  # process each character

        if c.islower():
            # Find where this cipher letter sits in cipher_lower,
            # then map that index back to plain_lower
            result += plain_lower[cipher_lower.index(c)]

        elif c.isupper():
            # Same logic for uppercase
            result += plain_upper[cipher_upper.index(c)]

        else:
            result += c  # non-alphabetic characters pass through unchanged

    return result


# ----------------------------------------------------------
# MAIN: Take input, encrypt, then decrypt to verify
# ----------------------------------------------------------
text = input("Enter text: ")

enc = encrypt(text)         # encrypt the input text
print("Encrypted:", enc)
print("Decrypted:", decrypt(enc))  # decrypt to confirm correctness
```
