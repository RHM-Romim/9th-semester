```python
# ============================================================
# LEHMANN'S PRIMALITY TEST (Euler's Criterion Based)
# ------------------------------------------------------------
# Lehmann's test is a PROBABILISTIC primality test.
# It does NOT guarantee a number is prime, but it can
# DEFINITELY say a number is COMPOSITE (not prime).
#
# WHY PROBABILISTIC?
#   Testing huge numbers (like RSA keys with 1024+ bits)
#   for primality deterministically is extremely slow.
#   Probabilistic tests are fast and accurate enough
#   for cryptographic use (error chance < (1/2)^t per run).
#
# CORE IDEA - Euler's Criterion:
#   For a prime p and any 'a' not divisible by p:
#       a^((p-1)/2) mod p  must equal  1  OR  p-1
#
#   If the result is ANYTHING ELSE -> p is definitely COMPOSITE.
#   If it always equals 1 or p-1 after t trials -> p is
#   'probably prime' (with higher confidence as t increases).
#
# EACH TRIAL:
#   1. Pick a random 'a' between 2 and p-2
#   2. Compute result = a^((p-1)/2) mod p
#   3. If result != 1 and result != p-1  =>  COMPOSITE (stop)
#   4. Otherwise repeat; after t trials  =>  PROBABLY PRIME
# ============================================================

import random


def lemann(p, t=10):
    """
    Lehmann's probabilistic primality test.
    p : the number to test
    t : number of random trials (higher t = higher confidence)
        Default is 10 trials -> error probability < (1/2)^10 = ~0.1%
    Returns True if p is probably prime, False if definitely composite.
    """

    # --- Edge case: handle small values directly ---
    # 2 is the only even prime; anything <= 1 is not prime
    if p <= 2:
        return p == 2   # True only if p is exactly 2

    # --- Run t independent trials ---
    for _ in range(t):   # repeat t times to increase confidence

        # STEP 1: Pick a random witness 'a' in range [2, p-2]
        # 'a' must not be 1 or p-1 to be a meaningful witness
        a = random.randint(2, p - 2)

        # STEP 2: Compute the exponent for Euler's Criterion
        # For a prime p: a^((p-1)/2) mod p should be 1 or p-1
        # (p-1)//2 uses integer division to get the floor
        e = (p - 1) // 2

        # STEP 3: Compute a^e mod p efficiently using Python's built-in pow
        # pow(base, exponent, modulus) uses fast modular exponentiation
        # This avoids computing huge numbers directly
        result = pow(a, e, p)   # result = a^((p-1)/2) mod p

        # STEP 4: Apply Euler's Criterion test
        # For a TRUE prime, result MUST be 1 or p-1 (which equals -1 mod p)
        # If result is anything else, p is DEFINITELY NOT prime -> return False
        if result != 1 and result != p - 1:
            return False   # COMPOSITE: found a witness that proves p is not prime

    # If all t trials passed, p is PROBABLY prime
    # (The more trials, the lower the chance of a false positive)
    return True


# ----------------------------------------------------------
# MAIN: Ask user for a number and test it
# ----------------------------------------------------------
num = int(input("Enter a number: "))

if lemann(num):
    print(f"{num} is prime.")       # probably prime (passed all t trials)
else:
    print(f"{num} is not prime.")   # definitely composite
```
